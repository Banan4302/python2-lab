Ćwiczenie 1: Series — tworzenie i operacje 

```python
import pandas as pd
import numpy as np
```

```python
# Series z listy — domyślny indeks (0, 1, 2, ...)
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

```python
# Klucze: 'Anna', 'Jan', 'Ewa', 'Marek', 'Kasia'
# Wartości: 5500, 7200, 4800, 9100, 6300
pensja = {'Anna': 5500,
          'Jan': 7200,
          'Ewa': 4800,
          'Marek': 9100,
          'Kasia': 6300}

pensje = pd.Series(pensja)
```

```python
# 1. Wyświetl imię osoby zarabiającej najwięcej (użyj idxmax())
# 2. Oblicz średnią pensję
# 3. Wyświetl imiona osób zarabiających powyżej średniej (filtrowanie boolean)
# 4. Policz ile osób zarabia poniżej 6000 (użyj len() lub .sum())
#1
print(pensje.idxmax())
print(pensje[pensje.idxmax()])
print(f'Najwięcej zarabia {pensje.idxmax()}.')
#2
print(pensje.mean())
#3
imiona = [k for k, v in pensja.items() if v > pensje.mean()]
for imie in imiona:
    print(f'{imiona}')
#4

# Sposób z len() - tworzymy tymczasową listę osób spełniających warunek
liczba_len = len([imie for imie in pensja if pensja[imie] < 6000])

# Sposób z sum() - sumujemy jedynki dla każdego trafienia (bardziej eleganckie)
liczba_sum = sum(1 for kwota in pensja.values() if kwota < 6000)

print(f"Liczba osób zarabiających poniżej 6000: {liczba_sum}")
```

```python
# 1. Oblicz pensje po podwyżce 10% dla wszystkich (pensje * 1.10)
# 2. Oblicz pensje po premii 500 zł dla wszystkich (pensje + 500)
# 3. Oblicz nową pensję = (pensja * 1.10) + 500
# 4. Wyświetl ranking nowych pensji od najwyższej (użyj sort_values)
#1
print(pensje * 1.10)
#2
print(pensje + 500)
#3
print(pensje *  1.10 + 500)
```

```python
#4
nowe_pensje = pensje *  1.10 + 500
ranking = nowe_pensje.sort_values(ascending= False)
print(ranking)
```

Ćwiczenie 2: DataFrame — tworzenie i atrybuty 

```python
# Dane o produktach sklepu internetowego
sklep = pd.DataFrame({
'produkt': ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Monitor'],
'cena': [3500, 1800, 2500, 350, 2200],
'stan_magazynowy': [45, 120, 200, 500, 75],
'kategoria': ['Komputery', 'Mobilne', 'Mobilne', 'Akcesoria', 'Komputery']
})
```

```python
# 1. Wyświetl kształt tabeli (shape)
print(sklep.shape)
# 2. Wyświetl nazwy kolumn (columns)
print(sklep.columns)
# 3. Wyświetl typy danych w każdej kolumnie (dtypes)
print(sklep.dtypes)
# 4. Wyświetl pierwsze 3 wiersze (head)
print(sklep.head(3))
```

```python
# 1. Dodaj kolumnę 'wartosc_magazynu' = cena * stan_magazynowy
sklep['wartość_magazynu'] = sklep['cena'] * sklep['stan_magazynowy']
print(sklep)
# 2. Dodaj kolumnę 'cena_z_vat' = cena * 1.23
sklep['cena_z_VAT'] = sklep['cena'] * 1.23
# 3. Wyświetl cały DataFrame z nowymi kolumnami
print(sklep)
```
