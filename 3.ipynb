```python
import numpy as np

# Dane wejściowe
miesiace_sprzedaz = np.array([120, 135, 98, 142, 167, 189, 201, 178, 156, 143, 198, 221])

print("--- Ćwiczenie 2: Część A ---")

# 1. Sprzedaż w styczniu (indeks 0)
print(f"1. Sprzedaż w styczniu: {miesiace_sprzedaz[0]}")

# 2. Sprzedaż w grudniu (ostatni element)
print(f"2. Sprzedaż w grudniu: {miesiace_sprzedaz[-1]}")

# 3. Sprzedaż w Q1 (styczeń-marzec, indeksy 0-2)
# stop=3, ponieważ indeks stop nie jest wliczany do wyniku
print(f"3. Sprzedaż w Q1: {miesiace_sprzedaz[0:3]}")

# 4. Sprzedaż w Q4 (październik-grudzień, ostatnie 3)
print(f"4. Sprzedaż w Q4: {miesiace_sprzedaz[-3:]}")

# 5. Sprzedaż co drugi miesiąc (styczeń, marzec, maj...)
print(f"5. Sprzedaż co drugi miesiąc: {miesiace_sprzedaz[::2]}")

# 6. Sprzedaż od tyłu (grudzień -> styczeń)
print(f"6. Sprzedaż od tyłu: {miesiace_sprzedaz[::-1]}")
```

```python
import numpy as np

# 1. Przygotowanie danych (Macierz 2D)
dane = np.array([
    [120, 145, 167],  # Produkt A
    [89, 102, 95],    # Produkt B
    [234, 267, 289],  # Produkt C
    [56, 78, 91],     # Produkt D
])

produkty = ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D']
kwartaly = ['Q1', 'Q2', 'Q3']

print("--- Ćwiczenie 2: Część B ---")

# 1. Sprzedaż Produktu A w Q2 (wiersz 0, kolumna 1)
print(f"1. Sprzedaż Produktu A w Q2: {dane[0, 1]}")

# 2. Cała sprzedaż Produktu C (wiersz 2, wszystkie kolumny)
print(f"2. Cała sprzedaż Produktu C: {dane[2, :]}")

# 3. Sprzedaż wszystkich produktów w Q1 (wszystkie wiersze, kolumna 0)
print(f"3. Sprzedaż wszystkich produktów w Q1: {dane[:, 0]}")

# 4. Sprzedaż Produktów B i C w Q2 i Q3 (wiersze 1-2, kolumny 1-2)
# Pamiętaj: stop (2 i 3) nie jest włączony w zakres!
print("4. Sprzedaż Produktów B i C w Q2 i Q3:")
print(dane[1:3, 1:3])
```

```python
import numpy as np

# --- Przygotowanie danych --- [cite: 112, 117]
nazwy = np.array(['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka', 
                  'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka'])
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])

# Zadanie 1: Podstawowe statystyki cenowe [cite: 120, 121]
print(f"Średnia cena: {ceny.mean()} zł") # 
print(f"Najdroższy produkt: {nazwy[ceny.argmax()]} ({ceny.max()} zł)") # 
print(f"Najtańszy produkt: {nazwy[ceny.argmin()]} ({ceny.min()} zł)") # 

# Zadanie 2: Wartość magazynu [cite: 123, 124]
wartosci_produktow = ceny * ilosci_magazyn # Mnożenie element-wise [cite: 125]
print(f"Łączna wartość magazynu: {wartosci_produktow.sum()} zł") # [cite: 127]

# Zadanie 3: Analiza cenowa [cite: 126]
ceny_rabat = ceny * 0.85 # Rabat 15% [cite: 128, 133]
ceny_vat = ceny * 1.23   # VAT 23% [cite: 129, 133]
powyzej_1000 = (ceny > 1000).sum() # Zliczanie produktów powyżej 1000 zł [cite: 131, 133]
print(f"Liczba produktów powyżej 1000 zł: {powyzej_1000}")

# Zadanie 4: Filtrowanie [cite: 134]
maska_oceny = oceny_klientow >= 4.5 # [cite: 135, 140]
print(f"Produkty z oceną >= 4.5: {nazwy[maska_oceny]}") # [cite: 136]

maska_tanie = ceny < 200 # [cite: 137]
print(f"Produkty poniżej 200 zł: {nazwy[maska_tanie]} (ilości: {ilosci_magazyn[maska_tanie]})") # [cite: 138]

# Zadanie 5: Analiza zapasów [cite: 141]
maska_niski_stan = ilosci_magazyn < 30 # [cite: 143]
print(f"Niski stan magazynowy: {nazwy[maska_niski_stan]}") # [cite: 146]
print(f"Wartość produktów z niskim stanem: {wartosci_produktow[maska_niski_stan].sum()} zł") # [cite: 144]

# Zadanie 6 (bonus): Analiza korelacji cena-ocena [cite: 148]
okazje = (oceny_klientow >= 4.5) & (ceny < 1000) # [cite: 149, 151]
print(f"Okazje (ocena >= 4.5 i cena < 1000): {nazwy[okazje]}")

srednia_wysoka = ceny[oceny_klientow >= 4.0].mean() # [cite: 150]
srednia_niska = ceny[oceny_klientow < 4.0].mean() # [cite: 150]
print(f"Średnia cena produktów z oceną >= 4.0: {srednia_wysoka:.2f} zł")
print(f"Średnia cena produktów z oceną < 4.0: {srednia_niska:.2f} zł")
```

```python
import numpy as np
import time

# --- Przygotowanie danych --- 
rozmiar = 1_000_000 # 
lista = list(range(rozmiar)) # 
tablica = np.array(lista) # 

# --- Test 1: Mnożenie przez 2 --- 
# Lista (Python)
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start # 

# NumPy
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start # 

print(f"=== Mnożenie x2 ({rozmiar:,} elementów) ===") # 
print(f"Lista: {czas_lista*1000:.1f} ms") # 
print(f"NumPy: {czas_numpy*1000:.1f} ms") # 
print(f"NumPy jest {czas_lista/czas_numpy:.0f}x szybszy!") # 

# --- Test 2: Suma --- 
# Lista (Python)
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start # 

# NumPy
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start # 

print(f"\n=== Suma ({rozmiar:,} elementów) ===") # 
print(f"Lista: {czas_lista*1000:.1f} ms") # 
print(f"NumPy: {czas_numpy*1000:.1f} ms") # 
print(f"NumPy jest {czas_lista/czas_numpy:.0f}x szybszy!") # 

# --- Test 3: Filtrowanie (elementy > 500 000) --- 
# Lista (Python)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start # 

# NumPy
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start # 

print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===") # 
print(f"Lista: {czas_lista*1000:.1f} ms") # 
print(f"NumPy: {czas_numpy*1000:.1f} ms") # 
print(f"NumPy jest {czas_lista/czas_numpy:.0f}x szybszy!") #
```
